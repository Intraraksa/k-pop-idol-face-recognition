```python
import cv2
import numpy as np
import dlib
import os
from statistics import mode
import pickle
```

```python
shape = dlib.shape_predictor('../shape_predictor_68_face_landmarks.dat')
detector = dlib.get_frontal_face_detector()
face_rec = dlib.face_recognition_model_v1('../dlib_face_recognition_resnet_model_v1.dat')
```

```python
face_desc = []
face_name = []
path = '../photo'
for folder in os.listdir(path):
    for file in os.listdir(os.path.join(path,folder)):
#         print(file)
        img = cv2.imread(path+'/'+folder+'/'+file)
#         img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        dets = detector(img,1)
        for k,d in enumerate(dets):
            sp = shape(img,d)
            face_dets = face_rec.compute_face_descriptor(img,sp,10)
            print(f'read : {file}')
            face_desc.append(face_dets)
            face_name.append(folder.split()[0])
            
pickle.dump((face_desc,face_name),open('trainset.pk','wb'))
```
